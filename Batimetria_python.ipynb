```python
import numpy as np
import netCDF4
import matplotlib
import matplotlib.pyplot as plt
import xarray as xr
from netCDF4 import Dataset , num2date




```

```python
def geo_idx(dd, dd_array):
    print(dd_array - dd)
    print(dd_array.shape, dd_array.dtype)
    geo_index = (np.abs(dd_array - dd)).argmin()
    return geo_index
```

```python
gebco = xr.open_dataset('/Users/Daniela/Documents/TRIATLAS/TP/gebco.nc')
lats = gebco.variables['lat'][:]
lons = gebco.variables['lon'][:]
print(lats.shape)
print(lons.shape)
```

```python
from matplotlib.colors import LinearSegmentedColormap

def custom_div_cmap(numcolors=11, name='custom_div_cmap',mincol='blue', midcol='white', maxcol='red'):
    
    cmap = LinearSegmentedColormap.from_list(name=name, colors =[mincol, midcol, maxcol], N=numcolors)

    return cmap
```

```python
blevels = [-6000, -5000, -4000, -3000, -2000, -1500, -1000, -500, -200, -100, 0]  #profundidades do meu local
N = len(blevels)-1
cmap2 = custom_div_cmap(N, mincol='DarkBlue', midcol='CornflowerBlue' ,maxcol='w')
cmap2.set_over('0.7') #nao colorir a terra 
```

```python
elevations = gebco.variables['elevation'][:]
print(elevations.shape)
print(elevations[1,1])
plt.figure(figsize=(12,6))
pc = plt.contourf(lons,lats,elevations, vmin=-6000, vmax=0, levels=blevels, cmap=cmap2, extend='both')
#plt.pcolormesh(lons,lats,elevations,cmap='jet',shading='auto')
cbar = plt.colorbar()
cbar.set_label('[Elevation]',rotation=0,labelpad=-30,y=1.09)
plt.xlabel('Longitude[°]')
plt.ylabel('Latitude[°]')
plt.title('Bathymetry')


#cbar_kwargs={"orientation": "horizontal", "label":"Bathymetry","shrink": 0.8}
plt.show()
```
